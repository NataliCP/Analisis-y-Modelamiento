```python
#Filtrar los municipios que tenemos
np.unique(municipiosgdf.MPIO_CNMBR)
```

```python
#filtrado por Departamento Cauca
municipiosfiltro = municipiosgdf[municipiosgdf['MPIO_CNMBR'] == 'VILLA RICA']
print (municipiosfiltro)
```

```python
# Filtrar los municipios de interés
municipios_interes = ['VILLA RICA', 'JAMUNDI', 'BUENOS AIRES', 'CALOTO', 'JAMBALO', 'CALDONO', 'SANTANDER DE QUILICHAO']
gdf_interes = municipiosgdf[municipiosgdf['MPIO_CNMBR'].isin(municipios_interes)]

# Imprimir los nombres de los municipios filtrados para verificar
print(gdf_interes['MPIO_CNMBR'])
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python

# Plotear los límites de los municipios
fig, ax = plt.subplots(figsize=(10, 10))
gdf_interes.plot(ax=ax, edgecolor='black', color='none')

# Añadir etiquetas a los municipios
for idx, row in gdf_interes.iterrows():
    plt.annotate(text=row['MPIO_CNMBR'], xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')

plt.title('Límites Geográficos de Santander de Quilichao y Municipios Vecinos')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
fig.savefig('limite_municipios.png', dpi=300, bbox_inches='tight')
plt.show()
```
